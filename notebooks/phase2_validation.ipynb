```python
# Phase 2 Checkpoint: LLM Provider Layer

from credit_analyzer.llm import get_provider

# 1. Factory + config
provider = get_provider()
print(f"Provider: {provider.model_name()}")
print(f"Available: {provider.is_available()}")

# 2. Real completion call
response = provider.complete(
    system_prompt="You are extracting information from a credit agreement.",
    user_prompt=(
        "Given this text: 'The Revolving Commitment is $50,000,000.' "
        "Extract: Facility type and amount."
    ),
    temperature=0.0,
)
print(f"\nResponse: {response.text}")
print(f"Tokens used: {response.tokens_used}")
print(f"Duration: {response.duration_seconds:.1f}s")
print(f"Model: {response.model}")
```
